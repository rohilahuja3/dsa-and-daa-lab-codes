        Consistency and Inconsistency

            1) r(A) = r[A,b] = no. of variable then it has unique solution
            2) r(A) = r[A,b] != no. of variable then it has infinitely many solutions
            3) r(A) != r[A,b] then system is inconsistent

```python
import numpy as np
A = np.array([[10,1,1],[1,10,1],[1,1,10]])
A1 = np.array([[12,12,12]]).T

Abig = np.column_stack((A,A1))
print(Abig,'\n')

c = np.linalg.matrix_rank(Abig)
print(c,'\n')

d = np.linalg.matrix_rank(A)
print(d,'\n')

if(c==d):
    print('The system is consistent',)
    s = np.linalg.solve(A,A1)
    print(s,'\n')
else:
    print('The system is inconsistent')
```
