```python
new_A = np.array([[0,1,1,-2],[1,2,-1,0],[2,4,1,-3],[1,-4,-7,-1]])
print(new_A,'\n')

from sympy import *
# new_A1 = Matrix([[0,1,1,-2],[1,2,-1,0],[2,4,1,-3],[1,-4,-7,-1]])   # to print the rref we first have to import sympy header file and declare the matrix this way
# new_A1 = Matrix(([[1,2,3],[4,5,6],[7,8,9]]))
new_A1 = Matrix(new_A)
print(new_A1.rref())

rank = np.linalg.matrix_rank(new_A)       # number of leading elements in ref of new_A as u can see new_A has 1 for 4 leading elements therefore rank = 4
print("Rank of A: ",rank)

trace = np.trace(new_A)
print("Trace of A: ",trace)

determinant = np.linalg.det(new_A)
print("Determinat of A: ",determinant,'\n')

inverse = np.linalg.inv(new_A)
print(inverse,'\n')

transpose = new_A.T
print(transpose,'\n')

power = np.linalg.matrix_power(new_A,3)
print(power)
```

        1) How to Coloumn stack and concatenate 2 matrices

```python
a1 = np.array([1,2,3])
b1 = np.array([4,5,6])
print(a1,'\n')
print(a1.shape,'\n')    #this is len of 1d array this is not a matrix and shape will print the number of elements in the list
print(np.concatenate((a1,b1)),'\n')   # concatenation can only work with 1d list elements not with matrices

a2 = np.array([[1],[2],[3]])   # this is a 3x1 matrix with 3 rows and 1 coloumn
b2 = np.array([[4,5,6]]).T     # this is a 1x3 matrix and to coloum stack it with a2 we need to have same dimension therefore we take transpose
print(a2,'\n')
print(a2.shape,'\n')
print(np.column_stack((a2,b2)),'\n')

a3 = np.array([[1,2,3]])      # this is a 1x3 matrix with 1 row and 3 coloumns
b3 = np.array([[4],[5],[6]]).T # this is 3x1 matrix and to coloum stack it dimension of a3 and b3 should be same therefore we take transpose
print(a3,'\n')
print(a3.shape,'\n')
print(np.column_stack((a3,b3)))
```

        4) how to solve a system of linear equations using LU decomposition
                a) Find Lu decomposition of matrix
                b) Solve the system of linear equations using LU decomposition 

        a) Find Lu decomposition of matrix from the equations given
            2x+y+z=4
            x+2y+z=4
            x+y+2z=4

```python
import numpy as np
from scipy.linalg import lu
A = np.array([[2,1,1,4],[1,2,1,4],[1,1,1,4]])
p,l,u = lu(A)
print(p,'\n')
print(l,'\n')
print(u,'\n')
```

        b) Solve using Lu decomposition
            2x+y+z=4
            x+2y+z=4
            x+y+2z=4

```python
import numpy as np
from scipy.linalg import lu_factor,lu_solve
A = np.array([[2,1,1],[1,2,1],[1,1,2]])   # 3x3 matrix
B = np.array([[4],[4],[4]])               # so this should be a 3x1 matrix
print(A.shape,B.shape,'\n')

lu,piv=lu_factor(A)
x=lu_solve((lu,piv),B)
print(x)
```
